Патч для исправления отсутствующих uid в проектах базы данных. Использовать единожды перед первым запуском синхронизатора.

```python
from src import PROJECT_FILE_NAME
from src.app import GeoOfficeSyncService

app = GeoOfficeSyncService()
```

```python
from src.utils.project_file_utils import ProjectFileUtils
from pony.orm import db_session
from src import PROJECT_FILE_NAME

with db_session:
    for project in app.database_service.get_all_projects():
        project_file_content = ProjectFileUtils.read_project_file(app.synchronization_service.projects_root_path / project.path / PROJECT_FILE_NAME)
        if project_file_content is None and (app.synchronization_service.projects_root_path / project.path).exists():
            project_file_content = ProjectFileUtils.set_uid(app.synchronization_service.projects_root_path / project.path / PROJECT_FILE_NAME)
        elif project_file_content:
            pass
        else:
            continue
        if not ProjectFileUtils.is_uid(project_file_content):
            uid = ProjectFileUtils.set_uid(app.synchronization_service.projects_root_path / project.path / PROJECT_FILE_NAME)
            project.uid = uid
            print(f"{project.path} + {uid}")
        else:
            project.uid = project_file_content
            print(f"{project.path} = {project_file_content}")
```
